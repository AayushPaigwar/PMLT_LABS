```python
import tensorflow as tf


tensor_a = tf.constant([[1, 2], [3, 4]], dtype=tf.float32)
tensor_b = tf.constant([[5, 6], [7, 8]], dtype=tf.float32)

print("Tensor A:")
print(tensor_a)
print("Tensor B:")
print(tensor_b)

```

```python
# Addition
tensor_add = tf.add(tensor_a, tensor_b)
print("\nAddition of Tensor A and Tensor B:")
print(tensor_add)
```

```python
# Subtraction
tensor_subtract = tf.subtract(tensor_a, tensor_b)
print("\nSubtraction of Tensor A and Tensor B:")
print(tensor_subtract)
```

```python
# Multiplication
tensor_multiply = tf.multiply(tensor_a, tensor_b)
print("\nMultiplication of Tensor A and Tensor B:")
print(tensor_multiply)
```

```python
# Division
tensor_divide = tf.divide(tensor_a, tensor_b)
print("\nDivision of Tensor A and Tensor B:")
print(tensor_divide)
```
